### Import Library and input file

```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv(os.getcwd()+'\input_file\Online_Retail_sample_10k.csv')
df.dropna(inplace=True)
df['CustomerID'] = df['CustomerID'].astype(int)
df.head()
```

### Create Function

This function will return multiple values, which will assign back to multiple column into dataframe

```python
def multi_column_ret():
    
    return ['val1','val2','val3']

multi_column_ret()
```

```python
## Parameter Details

# axis
#    * 0 or 'index': apply function to each column.
#    * 1 or 'columns': apply function to each row.

# result_type 
#    * 'expand' = will split the list and assign to respect 

df[['new_col1','new_col2','new_col3']] = df.apply(lambda X: multi_column_ret(), axis = 1 ,result_type = 'expand' )
df.head()
```

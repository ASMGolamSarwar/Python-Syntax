### Import Library and input file

```python
import pandas as pd
import numpy as np
import os
```

```python
#df = pd.read_excel(os.getcwd()+'\input_file\Online Retail.xlsx', sheet_name = 'Online Retail')

df = pd.read_csv(os.getcwd()+'\input_file\Online_Retail_sample_10k.csv')
df.dropna(inplace=True)
df['CustomerID'] = df['CustomerID'].astype(int)
df.head()
```

```python
df.info()
```

#### Aggregate on customerid level

Make sure the data has sorted accordingly

```python
df['YYYYMMDD'] = pd.to_datetime(df['InvoiceDate']).dt.strftime('%Y%m%d')
df = df.groupby(['CustomerID','YYYYMMDD']).agg({'UnitPrice':'sum'}).reset_index().sort_values(['CustomerID','YYYYMMDD'])
```

```python
df.head()
```

```python
## Get all the customer have transacted in multiple days
XX = df.groupby('CustomerID').count().sort_values('YYYYMMDD', ascending =False).reset_index()

XX[XX['YYYYMMDD']>1].head(5)
```

### Shift Rows Up | .shift(-1)

```python
df_shift = df[df['CustomerID'].isin(XX[XX['YYYYMMDD']>1]['CustomerID'])].copy()

df_shift.head()
```

```python
df_shift['next_day_visit'] = df_shift.groupby('CustomerID')['YYYYMMDD'].shift(-1)
df_shift.head()
```

### Shift Rows down | .shift(1)

```python
df_shift = df[df['CustomerID'].isin(XX[XX['YYYYMMDD']>1]['CustomerID'])].copy()

df_shift.head()
```

```python
df_shift['next_day_visit'] = df_shift.groupby('CustomerID')['YYYYMMDD'].shift(1)
df_shift.head()
```

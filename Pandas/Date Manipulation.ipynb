### Import Library and input file

```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_excel(os.getcwd()+'\input_file\Online Retail.xlsx', sheet_name = 'Online Retail')
df = df[['InvoiceDate']].head(100)
df.head()
```

```python
df.info()
```

### Extract Date/year/month/Day from timestamp/datetime

```python
df['date' ] = df['InvoiceDate'].dt.date
df['year' ] = df['InvoiceDate'].dt.year
df['month'] = df['InvoiceDate'].dt.month
df['day'  ] = df['InvoiceDate'].dt.day
df.head()
```

### Extract HH/MI/SS from timestamp/datetime

```python
df['HH'] = df['InvoiceDate'].dt.hour
df['MI'] = df['InvoiceDate'].dt.minute
df['SS'] = df['InvoiceDate'].dt.second
df.head()
```

### Convert timestamp/datetime into string

```python
df['YYYYMMDD'] = df['InvoiceDate'].dt.strftime('%Y%m%d')
df['HH24MISS'] = df['InvoiceDate'].dt.strftime('%H%M%S')
df.head()
```

### Convert string into date/datetime

```python
df.info()
```

look into the column type YYYYMMDD & HH24MISS 

```python
df['Conv_date'] = pd.to_datetime(df['YYYYMMDD'],format='%Y%m%d')
df.info()
df.head()
```

Conv_date data type is datetime 

```python
df['Conv_date_time'] = pd.to_datetime(df['YYYYMMDD']+df['HH24MISS'],format='%Y%m%d%H%M%S')
df.head()
```
